```python
from skimage import io
from matplotlib import pyplot as plt
```

```python
path = '/home/sabi/src/dip/data/owl.jpg'
```

```python
img = io.imread(path)
plt.imshow(img)
img.shape
```

```python
from skimage.transform import rescale, resize, downscale_local_mean
```

```python
# rescale img by 4 times (smaller)
img_rescale = rescale(img, 1.0/4.0, anti_aliasing=True)
plt.imshow(img_rescale)
img_rescale.shape
```

```python
# resize
img_resize = resize(img, (400,400))
plt.imshow(img_resize)
img_resize.shape
```

```python
# downscale by 4 times
img_downscale = downscale_local_mean(img, (4,4,3))
plt.imshow(img_downscale)
img_downscale.shape
```

```python
from skimage.filters import roberts, sobel, scharr, prewitt
```

```python
# import as gray
img = io.imread(path,1)
img.shape
```

```python
img_roberts = roberts(img)
plt.imshow(img_roberts,cmap=plt.cm.gray)
```

```python
img_sobel = sobel(img)
plt.imshow(img_sobel,cmap=plt.cm.gray)
```

```python
img_scharr = scharr(img)
plt.imshow(img_scharr,cmap=plt.cm.gray)
```

```python
img_prewitt = prewitt(img)
plt.imshow(img_prewitt,cmap=plt.cm.gray)
```

```python
from skimage.feature import canny
```

```python
img_canny = canny(img, sigma=3)
plt.imshow(img_canny)
```
