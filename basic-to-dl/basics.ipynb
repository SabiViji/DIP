```python
path = "/home/sabi/src/dip/data/watch.jpg"
```

```python
img = io.imread(path)
```

```python
plt.imshow(img)
```

# convert img to float type

```python
from skimage import img_as_float
```

```python
float_img = img_as_float(img)
plt.imshow(float_img)
```

```python
float_img.dtype
```

# Noise image

```python
noise_img = np.random.random([500,500])
plt.imshow(noise_img)
```

```python
noise_img.dtype
```

# Flip image

```python
imgLR = np.fliplr(img)
plt.imshow(imgLR)
```

```python
imgUD = np.flipud(img)
plt.imshow(imgUD)
```

# Rotate image

```python
from scipy import ndimage
```

```python
img45 = ndimage.rotate(img,45)
plt.imshow(img45)
```

```python
img75 = ndimage.rotate(img,75)
plt.imshow(img75)
```

# Fliter image

```python
img_uniform = ndimage.uniform_filter(img, size=9)
plt.imshow(img_uniform)
```

```python
img_gaussian = ndimage.gaussian_filter(img, sigma=3)
plt.imshow(img_gaussian)
```

```python
img_median = ndimage.median_filter(img, 9)
plt.imshow(img_median)
```

```python
img_sobel = ndimage.sobel(img, )
plt.imshow(img_sobel)
```
