```python
import numpy as np
import matplotlib.pylab as plt
%matplotlib inline
```

# Read image

```python
file = "data/owl.jpg"
img = plt.imread(file)
img.shape
```

```python
def display(im):
    """
    Helper function to display images
    """
    plt.imshow(im)
    plt.axis('off')
    plt.show()
```

```python
display(img)
```

# Image colors

```python
fig, axs = plt.subplots(nrows=1, ncols=3, figsize=(15,5))

for c, ax in zip(range(3), axs):
    tmp_img = np.zeros(img.shape, dtype="uint8")
    tmp_img[:,:,c] = img[:,:,c]
    ax.imshow(tmp_img)
    ax.set_axis_off()
```

```python
rgb_weights = [0.2989, 0.5870, 0.1140]
grayscale_image = np.dot(img[...,:3], rgb_weights)
#display(grayscale_image)
plt.imshow(grayscale_image, cmap=plt.get_cmap('gray'))
```

```python
# negative of the image
display(255 - img)
```

```python
# alternative method for grey
x = img[:,:,1]
plt.imshow(x,cmap=plt.get_cmap('gray'))
```

# Cropping

```python
img_crop = img[230:380,250:780,:].copy()
display(img_crop)
```

```python
# cut and paste img
img_copy = img.copy()
img_slice = img[250:350,250:450,:].copy()
img_copy[100:200,400:600,:] = img_slice
display(img_copy)
```

# Flip image

```python
img1 = img.copy()
display(np.flipud(img1))
display(np.fliplr(img1))
```
