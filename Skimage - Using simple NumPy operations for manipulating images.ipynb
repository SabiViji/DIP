```python
import numpy as np
from skimage import data, io
import matplotlib.pyplot as plt
```

```python
img = data.camera()
io.imshow(img)
```

```python
img[:10] = 0
io.imshow(img)
```

```python
mask = img < 87
io.imshow(mask)
```

```python
img[mask] = 255
io.imshow(img)
```

```python
inds_x = np.arange(len(img))
inds_y = (4*inds_x) % len(img)
img[inds_x, inds_y] = 0
io.imshow(img)
```

```python
l_x, l_y = img.shape[0], img.shape[1]
print(l_x,l_y)
```

```python
X, Y = np.ogrid[:l_x, :l_y]
print(X, Y)
```

```python
outer_disk_mask = (X-l_x / 2)**2 + (Y-l_y / 2)**2 > (l_x / 2)**2
io.imshow(outer_disk_mask)
```

```python
img[outer_disk_mask] = 0
io.imshow(img)
```

```python
plt.figure(figsize=(4, 4))
plt.imshow(img, cmap='gray')
plt.axis('off')
plt.show()
```
