The value of each grayscale pixel is calculated as the weighted sum of the curresponding red, green, blue pixels

W = 0.2125(R) + 0.7154(G) + 0.0721(B)

These weights are used by CRT phosphors as they represent human preception of red, green and blue then equal weights

```python
from skimage import data
from skimage.color import rgb2gray
import matplotlib.pyplot as plt
```

```python
img = data.astronaut()
gray = rgb2gray(img)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(8, 4))
ax = axes.ravel()

ax[0].imshow(img)
ax[0].set_title("Orginal")
ax[1].imshow(gray)
ax[1].set_title("Grayscale")

fig.thigt_layout()
plt.show()
```
